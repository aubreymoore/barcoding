This notebook looks up scientific names in GBIF.

If a name is found, the scrip checks for Guam occurrence records.

```python
import pygbif
import pandas as pd
import pprint
```

```python
df = pd.read_csv('repeat kyle.csv')
```

```python
for taxon in df.taxon.values: 
    print taxon
    taxon_dict = pygbif.species.name_lookup(q=taxon, limit=1)
    taxon_results = taxon_dict.get('results')
    if taxon_results:       
        pprint.pprint(taxon_results)
        #pprint.pprint(taxon_results, f)
    else:
        print '*** Taxon not found in GBIF'
        
    key = taxon_dict['results'][0].get('key', None)
    if key:
        print 'key = {}'.format(key)
        occurrence_dict = pygbif.occurrences.search(country='GU', taxonKey=key, limit=1)
        occurrence_results = occurrence_dict.get('results')
        if occurrence_results:
            pprint.pprint(occurrence_results)
        else:
            print '*** No GBIF occurrence records for Guam'
    else:
        print '*** key not found'
    print
```
