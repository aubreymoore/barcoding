This notebook looks up scientific names in the Micronesia Arthropods Database.  This database may be accessed online at http://guaminsects.net/mad/CheckList2.htm.

If a record is found for a given taxon, this does not necessarily mean that specimens were collected on Guam.

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
import pprint
```

```python
engine = create_engine('mysql+pymysql://readonlyguest:readonlypassword@mysql.guaminsects.net/mymad')
```

```python
df = pd.read_csv('repeat kyle.csv')
for taxon in df.taxon.values: 
    print taxon
    sql = 'select * from tblBiota where vchrFullName LIKE "{}%%"'.format(taxon)
    result = engine.execute(sql)
    if result.rowcount>0:
        for row in result:
            print row
    else:
        print '*** No record for this taxon in the Micronesia Arthopod Database'
    print
```
